```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# bringing in dataframe of csv file 

df = pd.read_csv("/Users/isabellale/Downloads/socialaddict.csv")

# setting the theme
sns.set_theme(style="darkgrid")

# calculating the mean scores for each country in csv
mean_scores = df.groupby("Country")["Avg_Daily_Usage_Hours"].mean().reset_index()

max_country= mean_scores.loc
# making sure labels are not squished
plt.figure(figsize=(11,15))

# creating the scatterplot
sns.scatterplot(x="Avg_Daily_Usage_Hours", y="Country", data = mean_scores)

# labeling the visualization
plt.title("Average Social Media Addiction Score by Country")
plt.ylabel("Country")
plt.xlabel("Average Hours Spent on Social Media Platforms")
plt.show()
```

Caption: This plot shows countries like the USA have the highest
average social media usage, while countries like Sweden have the lowest, 
highlighting the differences in social media habits between countries around the world

Explanation: 

The plot above represents the average daily social media usage (hours) for each country in the dataset. I chose a scatter plot, because it allowed each county to be displayed while also showing the variation of social media usage across the x-axis. As a result, it should be easy for readers to compare countries at a glance. From this plot, viewers should be able to see which country has the highest and lowest social media usage, and understand that social media habits can vary throughout the world. Overall, the plot is used to highlight differences and can be used to identify any patterns or outliers.
